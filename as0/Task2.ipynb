Firstly, import necessary libraries and datasets along with defining the number of inputs.

```python
import numpy as np
import pandas as pd

X = pd.read_csv("data/train_in.csv", header = None)
X[len(X.columns)] = 1
X = X.to_numpy()

y = pd.read_csv("data/train_out.csv", header = None).to_numpy().flatten()

w = np.random.randn(257)
print(w)
```

Next, let's define some functions

```python
def softmax(n):
    return np.exp(n ) / np.exp(n ).sum()
```

```python
def test(X, w):
    Z = np.dot(X, w)
    A = softmax(Z)
    return A
```

```python
def check_accuracy(y_pred, y):
    return np.sum(y == y_pred / len(y))
```

Next, let's train the model and check its accuracy.

```python
def fit(X, y, w):
    learning_rate = 0.1

    for i in range(1000):
        for j, x in enumerate(X):
            Z = np.dot(x, w)
            y_pred = softmax(Z)
            print(Z)
            w += x * (learning_rate * (y[j] - y_pred))
            
            
    
fit(X, y, w)
```

Time to test the model using the testing set

```python
X_test = pd.read_csv("data/test_in.csv", header = None)
X_test[len(X_test.columns)] = 1
X_test = X_test.to_numpy()

y_test = pd.read_csv("data/test_out.csv", header = None).to_numpy().flatten()

print(w)

y_test_pred = test(X_test, w)
test_accuracy = check_accuracy(y_test_pred, y_test)
print(test_accuracy)

print(y_test_pred)


```
